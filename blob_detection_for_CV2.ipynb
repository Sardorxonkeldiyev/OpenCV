```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def show(img):
    plt.imshow(img,cmap='gray')
    plt.show()
```

```python
def get_image(filename):
  img = cv2.imread(filename)
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
  return img, gray
```

```python
img, gray = get_image('/content/1676184824_grizly-club-p-razmetka-dorozhnaya-klipart-48.jpg')
```

```python
show(img)
```

```python
edge = cv2.Canny(img.copy(),400,450)
show(edge)
```
